# Functions

```python
def compute_HCF(a, b):
  """
  Computing HCF of two numbers      #Doc String
  """
  smaller_number = b if a > b else a
  hcf = 1
  for i in range(1, smaller_number+1):
    if (a%i == 0) & (b%i == 0):
      hcf = i
  return hcf

print("HCF of 988 and 1064 is", compute_HCF(988, 1064))
```

### Built in functions

```python
# abs() for absolute value of numbers
print(abs(-124), "is the absolute value of -124\n",          
      abs(124), "is the absolute value of 124")
      
# dir() for directory      
print("\nDirectory of funtions related to list datatype:\n", dir(list),"\n")    
      
#divmod() for returning a tuple having quotient and remainder      
print(divmod(9,2), "are the quotient and remainder for dividing 9 by 2")     

# isinstance()
print("\nChecking if (1, 2, 4) is a list or not:", isinstance((1, 2, 4), list))
```

```python
def find_negative_number(n):
  """
  This function returns the number only if it is negative      #Doc String
  """
  if n < 0:
    return n
```

```python
# filter(function, iterable)
numbers = range(-20, 20)
negative_numbers_list = list(filter(find_negative_number, numbers))
print(negative_numbers_list, "are the negative numbers")
```

```python
def square_of_number(num):
  """
  This function returns the square of a number
  """
  return num*num
```

```python
# map(function, iterable)
numbers = range(1, 5)
squared_numbers = list(map(square_of_number, numbers))
print(squared_numbers, "are the square of [1, 2, 3, 4] respectively")
```

```python
def multiply(x, y):
  """
  This function multiplies two numbers
  """
  return x*y
```

```python
# reduce()
from functools import reduce
product = reduce(multiply, numbers)
print("Multiplying all numbers in the 'numbers' list:", product)
```

### Anonymous/Lambda Functions

```python
y = lambda x : x*2
print(y(5))
```

#### Lambda functions using built in functions

```python
# Lambda using filter()
odd_list = list(filter(lambda x : (x%2 != 0), numbers))
print("Odd numbers:", odd_list)
```

```python
# Lambda using map()
new_list = list(map(lambda x : x*2, numbers))
print("Double of numbers:", new_list)
```

```python
# Lambda using reduce()
product_of_list = reduce(lambda x, y : x*y, numbers)
print("Product of all numbers:", product_of_list)
```

### Debugging using pdb

```python
import pdb

def seq(n):
  for i in range(n):
    pdb.set_trace()
    print(i)
  return

seq(5)
```
